## 2.2 Helper function

```python
# 增加DOW列
def add_dow_col(df):
    df_r = df.copy()
    df_r['dow'] = pd.to_datetime(df_r['visit_date']).dt.dayofweek
    return df_r

# 增加月份列
def add_month_col(df):
    df_r = df.copy()
    df_r['month'] = pd.to_datetime(df_r['visit_date']).dt.month
    return df_r
```
